```python
# If blood_pressure is a column in your dataframe
df[['systolic', 'diastolic']] = df['Blood Pressure'].str.split('/', expand=True)
df['systolic'] = pd.to_numeric(df['systolic'])
df['diastolic'] = pd.to_numeric(df['diastolic'])
df = df.drop(columns=['Blood Pressure'])
```

```python
cat_cols = [col for col in df.columns if df[col].dtype == 'object']
num_cols = [col for col in df.columns if df[col].dtype != 'object']
print("Categorical Columns:", cat_cols)
print("Numerical Columns:", num_cols)
```

```python
df[cat_cols].nunique()
```

```python
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None')
```

```python
df.isnull().sum() / df.shape[0] * 100
```

```python
df_processed = df.drop(columns=['Person ID'])
```

```python
df_encoded = pd.get_dummies(df_processed,
                            columns=['Gender', 'Occupation', 'BMI Category', 'Sleep Disorder'],
                            prefix=['Gender', 'Occupation', 'BMI', 'Disorder'],
                            drop_first=True,  # Use True for machine learning
                            dtype=int)
```
